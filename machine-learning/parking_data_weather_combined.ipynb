```python
import pandas as pd
```

```python
weather_df = pd.read_csv('sample-data/weather_data_10_years_preprocessed.csv')

weather_df.head()
```

```python
parking_df = pd.read_csv('sample-data/organized_parking_data_NEW.csv')

parking_df.head()
```

```python
# Merge the two DataFrames on 'Timestamp'
combined_df = pd.merge(parking_df, weather_df, on='Timestamp')

print(combined_df)
```

```python
combined_df.to_csv('combined_weather_parking.csv', index=False)
```
