```python
weather_df = pd.read_csv('./../sample-data/weather_data_10_years_preprocessed.csv')

weather_df.head()
```

```python
import pandas as pd

# Load the data from CSV
parking_df = pd.read_csv('./../sample-data/output.csv')

# Rename the 'Hour' column to 'Timestamp'
parking_df.rename(columns={'Hour': 'Timestamp'}, inplace=True)

# Display the first few rows of the DataFrame to verify the change
print(parking_df.head())
```

```python
# Merge the two DataFrames on 'Timestamp'
combined_df = pd.merge(parking_df, weather_df, on='Timestamp')

print(combined_df)
```

```python
combined_df.to_csv('./../sample-data/combined_weather_parking.csv', index=False)
```

```python

# Filter rows where timestamp is outside of the specified range
combined_df = combined_df[(combined_df['Timestamp'] < '2020-03-15') | (combined_df['Timestamp'] > '2021-09-07')]
# Drop duplicates, keeping the most recent entry for each timestamp
combined_df = combined_df.drop_duplicates(subset='Timestamp', keep='last')

combined_df.to_csv('./../sample-data/combined_weather_parking_WO2020_no_dups.csv', index=False)
```
